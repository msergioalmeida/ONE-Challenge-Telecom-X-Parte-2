# Escolha dos Modelos:

Para o problema de previsão de evasão de clientes, os modelos K-Nearest Neighbors (KNN) e Random Forest foram selecionados com base nas características dos dados e nos padrões observados nos gráficos exploratórios.

**K-Nearest Neighbors (KNN):**

O KNN é um modelo baseado em similaridade que pode identificar padrões de evasão ao comparar clientes com perfis semelhantes. É especialmente útil quando há concentração de evasores em determinadas faixas de tempo de contrato e faturamento, como observado nos gráficos. Ele é simples, intuitivo e eficaz em conjuntos de dados onde há estrutura espacial ou clusters visíveis, com as seguintes vantagens:
* Capta bem padrões locais e regiões de alta densidade de evasores.
* Útil para dados com padrões visuais e não lineares.
* Simples de implementar e interpretar.

**Random Forest Classifier (RF):**

O Random Forest é um modelo de aprendizado em conjunto baseado em múltiplas árvores de decisão. Ele é robusto, lida bem com variáveis correlacionadas e pode capturar interações complexas entre o tempo de contrato, faturamento mensal e total como mostrado nos boxplots.
Vantagens:
* Alta precisão e robustez.
* Capaz de lidar com dados desbalanceados (com class_weight).
* Fornece medidas claras de importância das variáveis.
* Lida bem com variáveis categóricas e numéricas sem necessidade de normalização.

**Conclusão:**

KNN e Random Forest se complementam bem: enquanto o KNN explora similaridade direta entre os clientes, o Random Forest captura interações mais complexas entre variáveis. Ambos são modelos eficazes para problemas com múltiplos fatores e padrões não lineares, como o de evasão de clientes em serviços de telecomunicações.

# 3.1- Modelo K-Nearest Neighbors (KNN)

Preparação

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix,ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split, cross_val_score, KFold, StratifiedKFold
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.under_sampling import RandomUnderSampler
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

X = df.drop("Evasao", axis=1)
y = df["Evasao"]

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=42)
```

Modelo KNN inicial

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)

print("KNN inicial:\n", classification_report(y_test, y_pred, digits=4))

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Não Evadiu', 'Evadiu'],
            yticklabels=['Não Evadiu', 'Evadiu'])
plt.xlabel('Predição')
plt.ylabel('Valor Real')
plt.title('Matriz de Confusão - KNN inicial')
plt.show()
```

Avaliação das variáveis mais importantes

```python
top_n_variaveis = list(range(1, len(df_importancias) + 1))

resultados = []
feature_counts = []
recalls = []

for n in top_n_variaveis:
    variaveis_top = df_importancias['Variavel'].head(n).tolist()

    X_train_top = X_train[variaveis_top]
    X_test_top = X_test[variaveis_top]

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_top)
    X_test_scaled = scaler.transform(X_test_top)

    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train_scaled, y_train)
    y_pred = knn.predict(X_test_scaled)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)

    resultados.append({
        'Variáveis': n,
        'Acurácia': round(acc, 4),
        'Precisão': round(prec, 4),
        'Recall': round(rec, 4),
        'F1-score': round(f1, 4)
    })

    feature_counts.append(n)
    recalls.append(rec)

df_resultados = pd.DataFrame(resultados)
print(df_resultados.to_string(index=False))

plt.plot(feature_counts, recalls, marker='o')
plt.xlabel("Quantidade de variáveis")
plt.ylabel("Recall")
plt.title("Recall vs. Quantidade de Variáveis")
plt.grid(True)
plt.xticks(range(0, max(feature_counts)+1, 5))
plt.ylim(0.45, max(recalls) + 0.01)
plt.show()
```

Função para mostrar métricas

```python
def avaliar_knn(titulo, y_pred):
    print(f"{titulo}:\n", classification_report(y_test, y_pred, digits=4))

    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Não Evadiu', 'Evadiu'],
                yticklabels=['Não Evadiu', 'Evadiu'])
    plt.xlabel('Predição')
    plt.ylabel('Valor Real')
    plt.title(f'Matriz de Confusão - {titulo}')
    plt.show()
```

KNN com as 10 variáveis mais relevantes

```python
top_features = df_importancias['Variavel'].head(10).tolist()

X_sel = df[top_features]

X_train, X_test, y_train, y_test = train_test_split(X_sel, y, stratify=y, random_state=42, test_size=0.3)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)

avaliar_knn("KNN com 10 variáveis principais", y_pred)
```

Validação cruzada com KFold

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(knn, X_train_scaled, y_train, cv=kf, scoring='accuracy')

print("Acurácia por fold:", np.round(scores, 4))
print("Acurácia média na validação cruzada:", round(scores.mean(), 4))
```

StratifiedKFold (SKF) e SMOTE na validação

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('knn', KNeighborsClassifier(n_neighbors=5))
])

scores = cross_val_score(pipeline, X_sel, y, cv=skf, scoring='recall')
print("Recall médio com SMOTE + StratifiedKFold:", round(scores.mean(), 4))
```

Pipeline com oversampling e validação cruzada

```python
pipeline_over = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('knn', KNeighborsClassifier())
])

param_grid_over = {
    'knn__n_neighbors': [3, 5, 7]
}

grid_over = GridSearchCV(pipeline_over, param_grid_over, cv=skf, scoring='recall')
grid_over.fit(X_train, y_train)

y_pred_test_over = grid_over.predict(X_test)

print("Melhores parâmetros (oversampling):", grid_over.best_params_)
print("Melhor recall na validação cruzada (oversampling):", grid_over.best_score_)

avaliar_knn("KNN pipeline com SMOTE", y_pred_test_over)
```

Pipeline com undersampling e validação cruzada

```python
pipeline_under = ImbPipeline([
    ('scaler', StandardScaler()),
    ('under', RandomUnderSampler(random_state=42)),
    ('knn', KNeighborsClassifier())
])

param_grid_under = {
    'knn__n_neighbors': [3, 5, 7, 9]
}

grid_under = GridSearchCV(pipeline_under, param_grid_under, cv=skf, scoring='recall')
grid_under.fit(X_train, y_train)

y_pred_test_under = grid_under.predict(X_test)

print("Melhores parâmetros (undersampling com KNN):", grid_under.best_params_)
print("Melhor recall na validação cruzada (undersampling com KNN):", grid_under.best_score_)

avaliar_knn("KNN pipeline com undersampling", y_pred_test_under)
```

```python
def avaliar_modelo(nome, y_true, y_pred):
    return {
        'Modelo': nome,
        'Acurácia': accuracy_score(y_true, y_pred),
        'Precisão': precision_score(y_true, y_pred),
        'Recall': recall_score(y_true, y_pred),
        'F1-Score': f1_score(y_true, y_pred)
    }

result_oversampling = avaliar_modelo("KNN + Oversampling", y_test, y_pred_test_over)
result_undersampling = avaliar_modelo("KNN + Undersampling", y_test, y_pred_test_under)

df_comparacao = pd.DataFrame([result_oversampling, result_undersampling])
print(df_comparacao.to_string(index=False))

cm_over = confusion_matrix(y_test, y_pred_test_over)
cm_under = confusion_matrix(y_test, y_pred_test_under)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

ConfusionMatrixDisplay(cm_over).plot(ax=axes[0], cmap="Blues", values_format='d')
axes[0].set_title("Matriz de Confusão - KNN com Oversampling")
axes[0].grid(False)

ConfusionMatrixDisplay(cm_under).plot(ax=axes[1], cmap="Greens", values_format='d')
axes[1].set_title("Matriz de Confusão - KNN com Undersampling")
axes[1].grid(False)

plt.tight_layout()
plt.show()
```

# Comparação de Modelos KNN - Oversampling e Undersampling:

A análise de desempenho dos modelos de machine learning para previsão de evasão de clientes foi conduzida utilizando as técnicas de oversampling e undersampling, com o objetivo de avaliar o impacto do balanceamento das classes nas métricas de avaliação, em especial no recall.

Analisando os resultados do KNN, observamos que o oversampling obteve recall de aproximadamente 0,708, enquanto o undersampling alcançou um recall superior, em torno de 0,772. Essa diferença é considerável e indica que, nesse caso, o undersampling foi mais eficaz para identificar clientes que realmente iriam evadir, atendendo melhor ao objetivo de maximizar a sensibilidade do modelo.

A precisão de ambos os métodos é bastante próxima — cerca de 0,495 para oversampling e 0,491 para undersampling —, o que sugere que a proporção de falsos positivos não muda de forma significativa entre as duas abordagens. A acurácia também permanece similar (0,730 no oversampling contra 0,727 no undersampling), mostrando que o ganho de recall com o undersampling não comprometeu o desempenho geral do modelo.

Portanto, para um cenário em que a detecção de potenciais evasões é prioridade, o KNN com undersampling se apresenta como a melhor opção. Ele entrega um recall claramente mais alto, mantendo valores de precisão e acurácia muito próximos aos do oversampling, o que o torna mais adequado para estratégias de retenção de clientes.

# 3.2- Modelo Random Forest Classifier (RF)

Preparação

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix,ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score, StratifiedKFold, GridSearchCV, train_test_split
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

X = df.drop("Evasao", axis=1)
y = df["Evasao"]

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=42)
```

Modelo RF inicial

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

Função para mostrar métricas

```python
def avaliar_rf(titulo, y_pred):
    print(f"{titulo}")
    print("Acurácia:", round(accuracy_score(y_test, y_pred), 4))
    print("Precisão:", round(precision_score(y_test, y_pred, zero_division=0), 4))
    print("Recall:", round(recall_score(y_test, y_pred, zero_division=0), 4))
    print("F1-score:", round(f1_score(y_test, y_pred, zero_division=0), 4))

    print("\nRelatório completo:\n")
    print(classification_report(y_test, y_pred, target_names=['Não Evadiu', 'Evadiu'], digits=4))

    cm = confusion_matrix(y_test, y_pred)

    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Não Evadiu', 'Evadiu'],
                yticklabels=['Não Evadiu', 'Evadiu'])
    plt.xlabel('Predição')
    plt.ylabel('Valor Real')
    plt.title(f'Matriz de Confusão - {titulo}')
    plt.show()
```

```python
avaliar_rf("Modelo RF inicial", y_pred)
```

Escolha da melhor quantidade de variáveis

```python
recalls = []
feature_counts = list(range(3, len(df_importancias) + 1))

for k in feature_counts:
    top_k_features = df_importancias['Variavel'].head(k).tolist()
    X_k = df[top_k_features]
    X_train, X_test, y_train, y_test = train_test_split(X_k, y, stratify=y, test_size=0.3, random_state=42)

    rf = RandomForestClassifier(random_state=42)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)

    recalls.append(recall_score(y_test, y_pred))

plt.plot(feature_counts, recalls, marker='o')
plt.xlabel("Quantidade de variáveis")
plt.ylabel("Recall")
plt.title("Recall vs. Quantidade de Variáveis")
plt.grid(True)
plt.show()
```

Seleção as variáveis mais importantes

```python
top_features = df_importancias['Variavel'].head(10).tolist()

X_top = df[top_features]

X_train, X_test, y_train, y_test = train_test_split(X_top, y, stratify=y, test_size=0.3, random_state=42)

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

avaliar_rf("Modelo RF 10 variáveis", y_pred)
```

Melhores parâmetros

```python
param_grid = {
    'max_depth': [5, 10, 15, None],
    'n_estimators': [100, 200],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

grid = GridSearchCV(RandomForestClassifier(random_state=42),
                    param_grid,
                    scoring='recall',
                    cv=skf)

grid.fit(X_top, y)
print("Melhores parâmetros:", grid.best_params_)
print("Melhor recall CV:", grid.best_score_)

melhor_modelo = grid.best_estimator_
y_pred = melhor_modelo.predict(X_test)

avaliar_rf("Modelo RF melhores parâmetros", y_pred)
```

Pipeline Undersampling com RandomUnderSampler

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

pipeline_rf_under = ImbPipeline([
    ('under', RandomUnderSampler(random_state=42)),
    ('rf', RandomForestClassifier(random_state=42))
])

param_grid_rf_under = {
    'rf__max_depth': [5, 10, 15, None],
    'rf__n_estimators': [100, 200],
    'rf__min_samples_split': [2, 5],
    'rf__min_samples_leaf': [1, 2]
}

grid_rf_under = GridSearchCV(pipeline_rf_under,
                             param_grid_rf_under,
                             scoring='recall',
                             cv=skf)

grid_rf_under.fit(X_train, y_train)

y_pred_rf_under = grid_rf_under.predict(X_test)

print("Melhores parâmetros (RF com undersampling):", grid_rf_under.best_params_)
print("Melhor recall CV:", grid_rf_under.best_score_)

avaliar_rf("RF com undersampling e melhores parâmetros", y_pred_rf_under)
```

A acurácia também é praticamente idêntica (0,745 para oversampling e 0,744 para undersampling), reforçando que a escolha entre as técnicas, nesse caso, deve realmente se apoiar na métrica de recall.

Portanto, se a prioridade for maximizar a detecção de clientes que irão evadir, mesmo que isso signifique lidar com alguns falsos positivos, o undersampling é a melhor escolha nesse conjunto de resultados. Ele entrega o maior recall sem prejuízo nas demais métricas, tornando-se mais eficiente para o seu objetivo de retenção de clientes.

A análise de desempenho mostra que o Random Forest com Undersampling apresentou o melhor Recall (0,8057), superando os demais modelos e garantindo maior capacidade de identificar clientes que realmente cancelam. O KNN também teve resultados competitivos, principalmente com Undersampling (Recall = 0,7718), porém com desempenho ligeiramente inferior ao Random Forest.
Resumo de Recall:
* **RF + Undersampling**: 0,8057 → Melhor desempenho.
* **RF + Oversampling**: 0,7807.
* **KNN + Undersampling**: 0,7718.
* **KNN + Oversampling**: 0,7076.
Isso indica que o uso de Undersampling favoreceu a detecção de evasores, possivelmente por equilibrar melhor as classes e reduzir viés para a maioria.

**3. Principais Fatores que Influenciam a Evasão**

Com base na análise de importância das variáveis e nos coeficientes dos modelos, identificam-se os seguintes pontos:
* **Meses de Contrato** (-1,2838) → Principal fator de retenção. Clientes com mais tempo de contrato têm menor probabilidade de cancelar, sugerindo fidelização ao longo do tempo.
* **Faturamento Total** (0,7031) → Surpreendentemente, clientes com maior faturamento total podem ter maior risco de evasão, possivelmente relacionado a contratos mais curtos com alto valor mensal.
* **Contrato Mensal** (0,4562) → Planos mensais apresentam maior risco de evasão do que contratos de longo prazo.
* **Serviço de Internet Fibra Óptica** (0,2505) → Apesar da qualidade esperada, clientes de fibra têm maior risco de evasão, podendo indicar concorrência agressiva nesse segmento.
* **Método de Pagamento Eletrônico** (0,2493) → Possível perfil de cliente mais sensível a preços e promoções.
* **Ausência de Suporte Técnico** (0,2306) e **Ausência de Segurança Online** (0,2238) → Clientes sem esses serviços adicionais demonstram maior probabilidade de cancelamento.
* **Faturamento Mensal** (0,1587) → Valores mais altos mensais estão associados a maior risco de evasão.

**4. Estratégias de Retenção Propostas**

Com base nos resultados e padrões identificados, sugerem-se as seguintes ações:
1. **Fidelização de contratos curtos**

2. **Ajuste de pacotes para alto faturamento mensal**
    * Oferecer revisões personalizadas para clientes com alto valor de faturamento mensal, incluindo pacotes mais flexíveis e competitivos.

5. **Monitoramento preventivo com base no modelo de predição**
